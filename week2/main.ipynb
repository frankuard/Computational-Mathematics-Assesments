### **Question 4**

**A positive integer is called a perfect number if it is** **equal to the sum of all of its divisors, including 1 but** **excluding the number itself. For example, 6 = 1 + 2 + 3.** **Write a function sumOfDivisors that takes a positive** **integer n and returns the sum of all its divisors (excluding n).**

**Answer:**

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1,n):
        if n%i ==0:
            total += i
    return total

sumOfDivisors(15)
```

### **Question 8**

**Write a function prodDigits() that inputs a number and returns the product of digits of that number.**

**Answer:**

```python
def prodDigits(num):
    product = 1
    digits = [int(i) for i in str(num)]
    for i in digits:
        product *=i
    print(product)
prodDigits(234)
```

### **Question 9**

**Write a function that will take in an unknown number of arguments and multiply all of them together and run the function for these 2 sets of numbers:**
**a. 1,2,3,4,5**
**b. 12,13,14**

**Answer:**

```python
def multiply_all(*args):
    result = 1
    for i in args:
        result *= i
    print(result)
    
multiply_all(1,2,3,4,5)
multiply_all(12,13,14)
```
