**3. Plot 20 values of x and y from the following functions.**
**f(x) = 5x^3+2x − 1**\
**f(x) = −2x^3 + x^2 + 100**\
**f(x) = 2πx + 20**

```python
def function1(x):
    result = 5*(x**3)+2*x-1
    return result

x_cord1 = []
y_cord1 = []
for i in range(20):
    x_cord1.append(i)
    y_cord1.append(function1(i))

plt.plot(x_cord1,y_cord1, marker="*", color="red")
plt.grid()
```

```python
def function2(x):
    result = -2*(x**3)+(x**2)+100
    return result

x_cord2 = []
y_cord2 = []
for i in range(20):
    x_cord2.append(i)
    y_cord2.append(function2(i))

plt.plot(x_cord2,y_cord2, marker="*", color="blue")
plt.grid()
```

```python
import math
def function3(x):
    result = (2*math.pi*x)+20
    return result

x_cord3 = []
y_cord3 = []
for i in range(20):
    x_cord3.append(i)
    y_cord3.append(function3(i))

plt.plot(x_cord3,y_cord3, marker="*", color="green")
plt.grid()
```

**4. Plot all the graph from Q no 3 on the same pane using subplot.**

```python
plt.subplot(2,2,1)
plt.plot(x_cord1,y_cord1, marker="*", color="red")

plt.subplot(2,2,2)
plt.plot(x_cord2,y_cord2, marker="*", color="green")

plt.subplot(2,2,3)
plt.plot(x_cord3,y_cord3, marker="*", color="yellow")

for ax in plt.gcf().axes:
    ax.set_xlabel("X-Axis")
    ax.set_ylabel("Y-Axis")
    ax.grid(True)
plt.tight_layout()
```
